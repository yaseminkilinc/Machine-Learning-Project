```python
import pandas as pd #Veri analizi
import numpy as np #Hesaplama
import requests #Web sayfalarına erişim
from bs4 import BeautifulSoup #HTML-XML ayrıştırma, işleme, web sayfalarından veri çekme
```

```python
url_list= [] # Trendyol ürünlerinin URL'lerini depolamak için boş bir liste oluşturuldu
prices_list=[] # Ürün fiyatlarını depolamak için boş bir liste oluşturuldu
propTitles=[] # Ürün özellik başlıklarını depolamak için boş bir liste oluşturuldu
propValues=[] # Ürün özellik değerlerini depolamak için boş bir liste oluşturuldu
mark_list=[] # Ürün markalarını depolamak için boş bir liste oluşturuldu

for i in range(1,30):
    url="https://www.trendyol.com/sr?wb=786%2C626%2C782%2C794%2C796%2C102900%2C105009%2C791%2C104971&wc=103623&sst=SCORE&pi=" + str(i)
    r=requests.get(url) # URL'den sayfanın içeriği çekildi
    source=BeautifulSoup(r.content,"lxml",from_encoding="utf-8") # İçerik BeautifulSoup kullanılarak parçalandı
    
    # Sayfadan ürün URL'leri bulundu ve işlendi
    urls=source.find_all("div",attrs={"class":"p-card-chldrn-cntnr"})
    for url in urls:
        url_freezer="https://www.trendyol.com/"+url.a.get("href")
        url_list.append(url_freezer) # Bulunan URL listesine eklendi
        print(url_freezer)
        
        # Ürünün detay sayfasından veriler çekildi
        r_freezer=requests.get(url_freezer)
        source_freezer=BeautifulSoup(r_freezer.content,"lxml")
        
        # Ürün özellikleri bulundu ve işlendi
        properties=source_freezer.find_all("li",attrs={"class":"attribute-item"})
        for prop in properties:
            prop_title=prop.find("span",attrs={"class":"attribute-label"}).text
            prop_value=prop.find("span",attrs={"class":"attribute-value"}).text
            propTitles.append(prop_title) # Özellik başlığı listesine eklendi
            propValues.append(prop_value) # Özellik değeri listesine eklendi
    
    # Ürün fiyatları bulundu ve işlendi
    prices=source.find_all("div",attrs={"class":"prc-box-dscntd"})
    for price in prices:
        prices_list.append(price.text) # Fiyat listesine eklendi
        print(price.text)
    
    # Ürün markaları bulundu ve işlendi
    marks=source.find_all("span",attrs={"class":"prdct-desc-cntnr-ttl"})
    for mark in marks:
        mark_list.append(mark.text) # Marka listesine eklendi
        print(mark.text)
    
print(str(len(url_list))+"Adet link bulundu.")
print(str(len(prices_list))+"Adet fiyat bulundu.")
print(str(len(propTitles))+"Adet özellik başlığı bulundu.")
print(str(len(propValues))+"Adet özellik verisi bulundu.")
```

```python
# Ürünlerin URL'lerini ve fiyatlarını içeren DataFrame oluşturuldu
df_urls=pd.DataFrame()
df_urls["urls"]=url_list # Ürün URL'leri sütunu oluşturuldu ve URL listesi ile dolduruldu
df_urls["prices"]=prices_list # Ürün fiyatları sütunu oluşturuldu ve fiyat listesi ile dolduruldu
df_urls["marks"]=mark_list # Ürün markaları sütunu oluşturuldu ve marka listesi ile dolduruldu
```

```python
freezers=len(url_list) # Toplamda kaç adet buzluk URL'si bulunduğunu belirlemek için URL listesinin uzunluğunu kullanarak bir değişken oluşturuldu
```

```python
columns=np.array(propTitles) #Özellik başlıklarını içeren propTitles listesi numpy dizisine (hesaplama) dönüştürüldü
columns=np.unique(columns)   #Tekrar eden özellik başlıkları kaldırıldı ve benzersiz başlıklar alındı
```

```python
df=pd.DataFrame(columns=columns) #Özellik başlıklarını sütun isimleri olarak kullanarak boş bir DataFrame oluşturuldu
df["urls"]=url_list #DataFrame'e ürün URL'lerini içeren sütun eklendi
df["prices"]=prices_list #DataFrame'e ürün fiyatlarını içeren sütun eklendi
df["marks"]=mark_list #DataFrame'e ürün markalarını içeren sütun eklendi
```

```python
df.head()
```

```python
for i in range(0,freezers):
    # URL'ye HTTP isteği gönderilerek sayfanın içeriği alındı ve BeautifulSoup ile parse edildi
    url=df['urls'].loc[i] 
    r=requests.get(url)
    source=BeautifulSoup(r.content,"lxml",from_encoding="utf-8")
    
    # Ürün özellikleri bulundu ve DataFrame'e eklendi
    properties=source.find_all("li",attrs={"class":"attribute-item"})
    for prop in properties:
        prop_title=prop.find("span",attrs={"class":"attribute-label"}).text
        prop_value=prop.find("span",attrs={"class":"attribute-value"}).text
        print(prop_title+prop_value) #Özellik başlığı ve değeri ekrana yazdırıldı
        df[prop_title].loc[i]=prop_value #DataFrame'e özellik başlığı ve değeri eklendi
```
