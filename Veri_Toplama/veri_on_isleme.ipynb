```python
import pandas as pd #Veri analizi
import numpy as np #Hesaplamalar
```

```python
data = pd.read_excel("veri_analizi.xlsx") #"veri_analizi.xlsx" dosyasından veriler okunarak bir DataFrame oluşturuldu
```

```python
#data DataFrame'inden seçilen sütunlar df DataFrame'ine kopyalandı
df['Markalar']=data['Markalar']
df['Buzluk_Tipi']=data['Buzluk_Tipi']
df['Derinlik']=data['Derinlik']
df['Dondurucu_Yeri']=data['Dondurucu_Yeri']
df['Dondurucu_Ozelligi']=data['Dondurucu_Ozelligi']
df['Enerji_Sinifi']=data['Enerji_Sinifi']
df['Garanti_Suresi']=data['Garanti_Suresi']
df['Tip']=data['Tip']
df['Toplam_Hacim']=data['Toplam_Hacim']
df['Ucretsiz_Kurulum_Var_Mi']=data['Ucretsiz_Kurulum_Var_Mi']
df['Fiyat']=data['Fiyat']
```

```python
#Belirli sütunlardaki boş değerler (boş string) NaN ile değiştirildi
df["Buzluk_Tipi"] = df["Buzluk_Tipi"].replace('', np.nan)
df["Derinlik"] = df["Derinlik"].replace('', np.nan)
df["Dondurucu_Yeri"] = df["Dondurucu_Yeri"].replace('', np.nan)
df["Dondurucu_Ozelligi"] = df["Dondurucu_Ozelligi"].replace('', np.nan)
df["Enerji_Sinifi"] = df["Enerji_Sinifi"].replace('', np.nan)
df["Tip"] = df["Tip"].replace('', np.nan)
df["Ucretsiz_Kurulum_Var_Mi"] = df["Ucretsiz_Kurulum_Var_Mi"].replace('', np.nan)

#Belirli sütunlardaki eksik değerler sütunun ortalaması ile dolduruldu
df["Derinlik"].fillna(df["Derinlik"].mean(), inplace=True)
df["Garanti_Suresi"].fillna(df["Garanti_Suresi"].mean(), inplace=True)
df["Toplam_Hacim"].fillna(df["Toplam_Hacim"].mean(), inplace=True)

# En çok tekrar eden string değeri bul
most_common_value = df["Buzluk_Tipi"].mode()[0]
# NaN değerleri en çok tekrar eden değer ile doldur
df["Buzluk_Tipi"].fillna(most_common_value, inplace=True)

# En çok tekrar eden string değeri bul
most_common_value = df["Dondurucu_Yeri"].mode()[0]
# NaN değerleri en çok tekrar eden değer ile doldur
df["Dondurucu_Yeri"].fillna(most_common_value, inplace=True)

# En çok tekrar eden string değerleri bul
most_common_islemci_tipi = df["Dondurucu_Ozelligi"].mode()[0]

# NaN değerleri en çok tekrar eden değerler ile doldur
df["Dondurucu_Ozelligi"].fillna(most_common_islemci_tipi, inplace=True)

# En çok tekrar eden string değeri bul
most_common_value = df["Enerji_Sinifi"].mode()[0]
# NaN değerleri en çok tekrar eden değer ile doldur
df["Enerji_Sinifi"].fillna(most_common_value, inplace=True)

# En çok tekrar eden string değeri bul
most_common_value = df["Tip"].mode()[0]
# NaN değerleri en çok tekrar eden değer ile doldur
df["Tip"].fillna(most_common_value, inplace=True)

# En çok tekrar eden string değeri bul
most_common_value = df["Ucretsiz_Kurulum_Var_Mi"].mode()[0]
# NaN değerleri en çok tekrar eden değer ile doldur
df["Ucretsiz_Kurulum_Var_Mi"].fillna(most_common_value, inplace=True)
```

```python
#Garanti yılı verilerini yuvarla
df['Garanti_Suresi'] = df['Garanti_Suresi'].apply(lambda x: int(x) if x == int(x) else int(x) + 1)
df["Garanti_Suresi"]=df.Garanti_Suresi.astype(float)
```
